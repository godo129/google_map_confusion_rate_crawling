```python
def append_dfs(datas,name,hosun,gu,dong,juso,lat,long) : 
    confusion_datas = []
    confusion_datas.extend([name,hosun,gu,dong,juso])
    
    for day in day_list :
        for time in time_list : 
            confusion_datas.append(datas[time][day])
            
    confusion_datas = confusion_datas + [lat,long]
        
    return confusion_datas
```
